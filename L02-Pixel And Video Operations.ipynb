<h4> Extract and save the frames</h4>

```python
vid = cv2.VideoCapture("images/nature.mp4")

#get the first frmaes and see if succesful
success , frame = vid.read()

#instiate count
count = 1
#Keep extracting frames as long as success is True
while success:
    cv2.imwrite(f"nature_images/ {count}.jpg", frame)
    success, frame = vid.read()
    count += 1
```

<h4>Capture videos using camera</h4>

```python
cap = cv2.VideoCapture(0)
```

```python
if cap.isOpened():
    print('Camera is present!')
else:
    print("Camera is not present")
```

```python
# Read until video is completed
while (cap.isOpened()):
    # capture frame-by-frame
    ret,frame = cap.read()
    if ret==True:
    #Display the resulting frame
        cv2.imshow('Frame',frame)
    #Press Q on keyboard to exist
        if cv2.waitKey(1) & 0xff == ord("q"):
            break

    else:
        break

#When everything done ,reease the video capture object
cap.release()

# Closes all the frames
cv2.destroyAllWindows()
```

<h4>Capture and store the objects</h4>

```python
#Create an Object to read
# from camera 
video = cv2.VideoCapture(0)
```

```python
# we need to set resolutions. so , convert them from float to integer.
frame_width = int (video.get(3))
frame_height = int(video.get(4))

size = (frame_width, frame_height)
```

```python
size
```

```python
# Below Video  Writer object create a frame of above defined the outputs 
# is stored in 'filename.avi'file
result = cv2.VideoWriter('myvid.avi', cv2.VideoWriter_fourcc(*'MJPG'),30,size)
```

```python
while(True):
    ret, frame = video.read()
    if ret == True:
        # Write the frame into the file 'filename.avi'
        result.write(frame)
        #Dispaly the frame saved in the file
        cv2.imshow('Frame',frame)
        # Press S on keyboard to stop the process
        if cv2.waitkey(1) & 0xff == ord('s'):
            break

    # break the loop
    else:
        break

cv2.destroyAllWindows()
video.release()
```
