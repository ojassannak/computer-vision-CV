<h4>Basic Histogram</h4>

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import  seaborn as sns
```

```python
marks = np.array ([56,66,56,87,67,62,28,65,48,89,82,66,59,52,60,63,46,67,88])
```

```python
plt.title('Histogram of marks')
plt.xlabel('Range of Marks')
plt.xticks(range(0,101,10))
plt.grid()
plt.ylabel('count')
plt.hist(marks,bins= range(0,101,10), color='Red');
```

```python
sns.histplot(marks, kde=True)
```

<h4>Color image  histogram</h4>

```python
img = cv2.imread("images/lenna.png")
```

```python
b, g, r = cv2.split(img)
```

```python
plt.figure(figsize=(16,9))
plt.title("Color Image Histogram")
plt.subplot(1,3,1)
plt.xticks(range(0,256,10))
plt.hist(b.flatten(), bins=range(0,256), color='b') 
plt.subplot(1,3,2)
plt.xticks(range(0,256,10))
plt.hist(g.flatten(), bins=range(0,256), color='g') 
plt.subplot(1,3,3)
plt.xticks(range(0,256,10))
plt.hist(r.flatten(), bins=range(0,256), color='r') 
plt.show()
```

<h4>Stretching andd Shrinkig</h4>

```python
image = cv2.imread("images/lenna.png")

new_width, new_height = 512,200

stretched_image = cv2.resize(image, dsize= (new_width, new_height),
                             interpolation = cv2.INTER_CUBIC)

cv2.imshow("stetched Image",stretched_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# Load the image
image = cv2.imread("images/lenna.png")

# Degine new dimensions
new_width, new_height = 200,512
# Resize the image to the new dimensions using cubic interpolation
stretched_image = cv2.resize(image, dsize= (new_width, new_height),
                             interpolation = cv2.INTER_CUBIC)

cv2.imshow("stetched Image",stretched_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h4>Histogram Equalization</h4>

```python
#import numpy
import numpy as np

# read image using method
img = cv2.imread('images/lenna.png')
gray= cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# creating a Histogram Equalization of as image using cv2. eualizeHist()
equ = cv2.equalizeHist(gray)

# show image inputs vs output
cv2.imshow('Equalized image', equ)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
