```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
image = cv2.imread('images/lenna.png',0)
```

```python
# Sobel parameters: ksize, dx, dy , ddepth
sobelx = cv2.Sobel(image,cv2.CV_64F,1,0, ksize=3) # Sobel for horizonatl edges
sobely = cv2.Sobel(image,cv2.CV_64F,0,1, ksize=3) # Sobel for vertical edges

# Combine the sobel immages(optional)
combined_sobel =cv2.add(sobelx,sobely)
```

```python
cv2.imshow("Sobel X", sobelx)
cv2.imshow("Sobel Y", sobely)
cv2.imshow("Combined", combined_sobel)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h4>Scharr Operator</h4>

```python
import cv2
# X gradient Scharr operator
fieldx = cv2.Scharr(image, cv2.CV_32F, 1, 0)/15.34

# Y gradient Scharr operator
fieldy = cv2.Scharr(image, cv2.CV_32F,0,1)/15.34

combined_scharr = cv2.add(fieldx, fieldy)
```

```python
cv2.imshow("Scharr X", fieldx)
cv2.imshow("Scharr Y", fieldy)
cv2.imshow("Combined Scharr",combined_scharr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h4> Canny Edge Detection Algorithm</h4>

```python
import cv2

# load the images
img= cv2.imread('images/rgy.jpg',0) # Load as grayscale

# Apply Canny edge detection
edges = cv2.Canny(img,50,150) # Adjust thresholds if needed

# Display results
cv2.imshow("Original Image",img)
cv2.imshow("Canny Edge Detetction",edges)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

# load the images
img= cv2.imread('images/lenna.png',0) # Load as grayscale

# Apply Canny edge detection
edges = cv2.Canny(img,50,180) # Adjust thresholds if needed

# Display results
plt.figure(figsize=(12,9))
plt.subplot(121)
plt.imshow(img, cmap="gray")
plt.subplot(122)
plt.imshow(edges, cmap = "gray")
plt.show()
```

```python
import cv2

# load the images
img= cv2.imread('images/lenna.png',0) # Load as grayscale

blurred = cv2.GaussianBlur(img, (5,5),0)

# Apply Canny edge detection
edges = cv2.Canny(blurred,50,120) # Adjust thresholds if needed

# Display results
plt.figure(figsize=(10,4))
plt.subplot(121)
plt.title("Original Image")
plt.imshow(img, cmap = "gray")
plt.subplot(122)
plt.title("Edges")
plt.imshow(edges, cmap="gray")
plt.show()
```

<h4>Countours Detection</h4>

```python
img  = cv2.imread('images/rgy.jpg')

# Convert to grayscale (often better for contour detection
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

#Creates a binary image where pixels aboce 127  are set to 255 (white)
#and others to 0(black)
thresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)[1]

# cv2find contours find contours in the binary image
# cv2.RETR_TREE retrieves all contours and creates a hierarchy tree
# cv2 CHAN_APPROX_SIMPLE approximates cntours to savemoney
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE,
                                       cv2.CHAIN_APPROX_SIMPLE)
# -1 draws all contours
# (0,255,0) is the color (green)
# 2 is the thickness of the lines
cv2.drawContours(img, contours,-1,(0,255,0),2) # draw alll contours in green

# Display the results
cv2.imshow('Original Image',img)
cv2.imshow('Thresholded Image',thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
