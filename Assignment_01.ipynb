<h4>2. Show the image in the matrix of 5 x 5 with following colormaps:</h4>

```python
img = cv2.imread("images/sachin.jpg")

plt.figure(figsize =(10,10))
plt.subplot(4,4,1)
plt.title('coolwarm')
plt.imshow(img, cmap ='coolwarm')

plt.subplot(4,4,2)
plt.title('rainbow')
plt.imshow(img, cmap ='rainbow')

plt.subplot(4,4,3)
plt.title('tab10')
plt.imshow(img, cmap ='tab10')

plt.subplot(4,4,4)
plt.title('seismic')
plt.imshow(img, cmap ='seismic')

plt.subplot(4,4,5)
plt.title('turbo')
plt.imshow(img, cmap ='turbo')

plt.subplot(4,4,6)
plt.title('Reds')
plt.imshow(img, cmap ='Reds')

plt.subplot(4,4,7)
plt.title('YlOrBr')
plt.imshow(img, cmap ='YlOrBr')
```
