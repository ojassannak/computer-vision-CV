<h4>Drawing a 3D line</h4>

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Define start and end point of the line in 3d space
start_point = (10,22,33)
end_point = (44,52,67)

#Create a 3d plot
fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")

# plot the line 
ax.plot([start_point[0], end_point[0]],
        [start_point[1], end_point[1]],
        [start_point[2], end_point[2]], 'r-')

ax.plot([10,40],[55,35],[55,30],'g')

ax.set_xlabel('X')
ax.set_ylabel("Y")
ax.set_zlabel("z")
ax.set_title("3D line")

plt.show()
```

```python
import matplotlib.pyplot as plt

#Define rectangle coordinates
x_min,y_min, x_max, y_max = 1,2,4,7

#Create the rectangle patch
rectangle = plt.Rectangle((x_min, y_min), x_max - x_min, y_max - y_min,
                          color ='red' , alpha = 0.1)

#Add the rectangle to a plot and customize labels/title
plt.plot([x_min,x_max],[y_min, y_min], 'b-', label= 'X-axis')
plt.plot([x_min,x_min],[y_min, y_max], 'g-', label= 'Y-axis')
plt.gca().add_patch(rectangle)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Rectangle Plot")
plt.legend()


# show the plot
plt.show()

```

<h4>Drawing the Ellipse</h4>

```python
#load an image
image = cv2.imread("images/lenna.png")

#Define center cordinates, radius and angle
center = (150,100) # (x, y)
axes= (100,50) # (major radius, minor radius)
angle = 45 # angle of rotation in degrees

#Draw a green ellipse with thickness of 3
cv2.ellipse(image,center,axes, angle, 0, 360,(0,255,0),3)

#Display the image with the ellipse
cv2.imshow("Image With Ellipse", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h4>Ellipse using matplotlib</h4>

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# Define enter coordinates , raddi and angle
center = (0,0)
width = 8
height =5
angle= 45

 #Create the eellipse object
ellipse = Ellipse(xy= center, width= width, height= height, angle= angle, color='blue')

# Add the ellipse to the plot
fig,ax = plt.subplots()
ax.add_patch(ellipse)
ax.set_aspect('equal') # Maintain qual aspect ratio for correct shape

# Set limits slighter wider than the ellipse
ax.set_xlim(-width -width/5, width + width/5)
ax.set_ylim(-height - height/5,height + height/5)

# add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Ellipse Plot')

# Show the plot
plt.show()
```

<h4>Drwaing circle</h4>

```python
#load an image
image = cv2.imread("images/lenna.png")

#Define center cordinates, radius and angle
center = (250,250) # (x, y)
radius= 250 # (major radius, minor radius)

#Draw a green ellipse with thickness of 3

for radius in range(10,400,5):
    cv2.circle(image,center,radius,(radius,127,0),2)
    
#Display the image with the ellipse
cv2.imshow("Image With Ellipse", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
