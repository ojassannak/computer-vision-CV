<h4>Drawing a 3D line</h4>

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

# Define start and end point of the line in 3d space
start_point = (10,22,33)
end_point = (44,52,67)

#Create a 3d plot
fig = plt.figure()
ax = fig.add_subplot(111, projection = "3d")

# plot the line 
ax.plot([start_point[0], end_point[0]],
        [start_point[1], end_point[1]],
        [start_point[2], end_point[2]], 'r-')

ax.plot([10,40],[55,35],[55,30],'g')

ax.set_xlabel('X')
ax.set_ylabel("Y")
ax.set_zlabel("z")
ax.set_title("3D line")

plt.show()
```

<h4>Ellipse using matplotlib</h4>

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Ellipse

# Define enter coordinates , raddi and angle
center = (0,0)
width = 8
height =5
angle= 45

 #Create the eellipse object
ellipse = Ellipse(xy= center, width= width, height= height, angle= angle, color='blue')

# Add the ellipse to the plot
fig,ax = plt.subplots()
ax.add_patch(ellipse)
ax.set_aspect('equal') # Maintain qual aspect ratio for correct shape

# Set limits slighter wider than the ellipse
ax.set_xlim(-width -width/5, width + width/5)
ax.set_ylim(-height - height/5,height + height/5)

# add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_title('Ellipse Plot')

# Show the plot
plt.show()
```

```python
#create the main window
root = Tk()
root.title("Drawing Canvas")
root.geometry('1000x800')

#Create the Canvas widget
canvas = Canvas(root, width=1000, height=800, bg='white')
canvas.pack()

# Define the Drawing Functions
def draw_line(event):
    x1,y1 = event.x, event.y
    canvas.create_line(x1,y1, x1+2,y1+2, width=10)

def draw_circle(event):
    x1,y1 = event.x, event.y
    canvas.create_oval(x1 - 5, y1 -5, x1+5, y1+5, fill= "yellow")

# bind mouse event to the draw the functions
canvas.bind("<Button-3>", draw_circle)
canvas.bind("<B1-Motion>",draw_line)

# Start the main loop
root.mainloop()
```

<h4>Mouse Operations</h4>

```python
import cv2

# Load a Image
img = cv2.imread("images/lenna.png")

# Define the callback Function
def draw_circle(event,x , y, flags, param):
    if event == cv2.EVENT_LBUTTONDOWN:
        cv2.circle(img,(x,y),10,(0,255,0), -1)  # Draw agreen cirle

#Bind the callback function
cv2.namedWindow("image")
cv2.setMouseCallback("image",draw_circle)

# Sho the image and ait for events
while True:
    cv2.imshow("image",img)
    if cv2.waitKey(1) == ord("q"):
        cv2.imwrite("New_Img.png", img)
        break
        
cv2.destroyAllWindows()
```

 <h4> Image Annotations</h4>

```python
import cv2
# Define text and rectangel properties
text= "Lenna"
font_face = cv2.FONT_HERSHEY_COMPLEX
font_scale =1.0
color = (0,255,145) # Green
thickness = 2

#calculate text width and heights
text_size, baseline = cv2.getTextSize(text, font_face, font_scale, thickness)

#Define text and rectangle positions
text_origin = (100, 100)
box_origin = (text_origin[0] - 10, text_origin[1] -baseline-20)
box_width, box_height = text_size[0] + 20, text_size[1] +20

image = cv2.imread("images/lenna.png")

# Draw the rectangele and text
cv2.rectangle(image, box_origin,(box_origin[0]+ box_width,
                                box_origin[1]+ box_height), color, thickness)
cv2.putText(image, text,text_origin, font_face, font_scale, color, thickness)

cv2.imshow("Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
