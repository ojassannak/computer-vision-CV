<h4>Averaging Kernel</h4>

```python
import cv2
import numpy as np

# load the image 
image = cv2.imread("images/lenna.png")

kernel = np.ones((5,5), np.float32) / 25

smoothed_image = cv2.filter2D(image, -1, kernel)

cv2.imshow("Orignial", image)
cv2.imshow("Smoothed", smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
kernel
```

```python
final_image = np.vstack([image, smoothed_image])
```

```python
cv2.imshow("Final",final_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h4>Gaussian Kernel</h4>

```python
kernel_size = 7 # Adjust as needed
sigma = 0.85 * ((kernel_size - 1) * 0.5 -1) + 2.35 # Gaussian standard deviation
kernel = cv2.getGaussianKernel(kernel_size, sigma)
```

```python
kernel
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.plot(range(7), kernel)
```

```python
smoothed_image = cv2.filter2D(image, -1, kernel)

cv2.imshow("Orignial", image)
cv2.imshow("Smoothed", smoothed_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
final_image = np.vstack([image, smoothed_image])

cv2.imshow("Final", final_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
avg_kernel = np.ones((5,5), np.float32) / 25
sigma = 0.85 * ((7 - 1) * 0.5 -1) + 2.35 # Gaussian standard deviation
gauss_kernel = cv2.getGaussianKernel(7, sigma)
```

```python
avg_image = cv2.filter2D(image, -1, avg_kernel)
gauss_image = cv2.filter2D(image, -1, gauss_kernel)
```

```python
comp_image = np.hstack([image, avg_image, gauss_image])
```

```python
cv2.imshow("Comapare", comp_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

<h4>Median Filter</h4>

```python
img = cv2.imread("images/lenna.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
kernal_size = 1
median = cv2.medianBlur(gray,kernel_size)
```

```python
median.shape
```

```python
plt.figure(figsize=(11,9))
plt.subplot(1,2,1)
plt.title("Original Image")
plt.imshow(gray, cmap='gray')
plt.subplot(1,2,2)
plt.title("Median Blured Image")
plt.imshow(median, cmap='gray')
```

```python
median3 = cv2.medianBlur(gray,3)
median5 = cv2.medianBlur(gray,5)
median7 = cv2.medianBlur(gray,7)
median9 = cv2.medianBlur(gray,9)
median11 = cv2.medianBlur(gray,11)
median13 = cv2.medianBlur(gray,13)
median15 = cv2.medianBlur(gray,15)
```

```python
plt.figure(figsize=(10,9))
plt.subplot(2,4,1)
plt.title("Original")
plt.imshow(gray, cmap='gray')
plt.subplot(2,4,2)
plt.title("Median -3")
plt.imshow(median3, cmap='gray')
plt.subplot(2,4,3)
plt.title("Median -5")
plt.imshow(median5, cmap='gray')
plt.subplot(2,4,4)
plt.title("Median -7")
plt.imshow(median7, cmap='gray')
plt.subplot(2,4,5)
plt.title("Median -9")
plt.imshow(median9, cmap='gray')
plt.subplot(2,4,6)
plt.title("Median -11")
plt.imshow(median11, cmap='gray')
plt.subplot(2,4,7)
plt.title("Median -13")
plt.imshow(median13, cmap='gray')
plt.subplot(2,4,8)
plt.title("Median -15")
plt.imshow(median15, cmap='gray')
plt.tight_layout()
```
